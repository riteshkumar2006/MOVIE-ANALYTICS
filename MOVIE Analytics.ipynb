```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv("IMDB-Movie-Data.csv")
```

```python
print("First 5 Rows :")
df.head()
```

```python
print("Last 5 rows")
df.tail()
```

```python
df.info()
df.describe()
```

```python
df=df.dropna(subset=['Rank', 'Title', 'Genre', 'Description', 'Director', 'Actors', 'Year',
       'Runtime (Minutes)', 'Rating', 'Votes', 'Revenue (Millions)',
       'Metascore'])
df.shape[0]
df.reset_index()
```

```python
## Exploratory Data Analysis........................................
```

```python
##Top 10 Movies by Rating
```

```python
top_movies=df.sort_values(by='Rating',ascending=False)[['Title','Rating']].head(10)
```

```python
print(top_movies)
```

```python
 ##Top 5 Highest Revenue Movies
```

```python
most_revenue=df.sort_values(by='Revenue (Millions)',ascending=False)
most_revenue.head()
```

```python
#year wise total revenue
```

```python
total_revenue_year = df.groupby("Year")["Revenue (Millions)"].sum()
print(total_revenue_year)
```

```python
total_revenue_year.plot(kind='bar',cmap='Pastel1',edgecolor='black')
plt.xlabel("Years")
plt.ylabel("Rupees (in Millions)")
plt.title("Yearwise Total Revenue")
plt.tight_layout()
plt.grid(alpha=0.9,axis='y')
```

```python
 ##  Movies Released per Year 
```

```python
movie_per_year=df.groupby("Year")["Title"].nunique()
print(movie_per_year)
```

```python
movie_per_year.plot(kind='barh',cmap='tab10')
plt.xlabel("Numbers of Movies Release Per Year")
plt.title("Movie Releases per year")
plt.tight_layout()
plt.grid()
```

```python
#average rating per year
```

```python
avg_rating=df.groupby("Year")["Rating"].mean()
print(avg_rating)
```

```python
avg_rating.plot(kind='bar',color="mediumseagreen",edgecolor='grey')
plt.xlabel("Year")
plt.ylabel("Rating")
plt.title("Averrage Rating Per Year")
plt.tight_layout()
```

```python
## Top 10 Genres by Average Revenue
```

```python
avg_rev_genres=df.groupby("Genre")["Revenue (Millions)"].mean().sort_values(ascending=False).head(10)
print(avg_rev_genres)
```

```python
colors = plt.cm.Set2(np.linspace(0, 1, 10))
avg_rev_genres.plot(kind="bar",color=colors,edgecolor="grey")
plt.xlabel("Genre")
plt.ylabel("Revenue (Millions)")
plt.title("Top 10 Genres by Average Revenue")
plt.tight_layout()
plt.grid(alpha=0.6,linestyle='--')
```

```python
##Top 10 Directors by revenue
```

```python
top_directors_revenue=df.sort_values(by='Revenue (Millions)',ascending=False)[['Director','Revenue (Millions)']].head(10)
print(top_directors_revenue)
```

```python
#Line Plot: Year vs Avgerage Revenue
avgrage_rev_year = df.groupby('Year')['Revenue (Millions)'].mean()
plt.plot(avgrage_rev_year.index,avgrage_rev_year.values,marker="D")
plt.style.use("dark_background")  
plt.title('Avg Revenue Per Year')
plt.xlabel('Year')
plt.ylabel('Revenue (Millions)')
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
```

```python
# Movie Share by Genre
top_genres=df["Genre"].value_counts().head(5)
colors = ['#98f078','#9c6436','#5f71e3','#d1db65','red']
plt.pie(top_genres,labels=top_genres.index,colors=colors,autopct='%1.1f%%',shadow=True,explode=[0.2,0,0,0,0])
plt.title(' Top 5 Genres by Number of Movies')
plt.tight_layout()
```
